```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi=np.linspace(-1,1,200)
cos=np.cos(np.pi*phi)
```

Asymmetric SQUID derivation:

$I=I_{C_1}\sin{\delta_1}+I_{C_2}\sin{\delta_2}$

Given $\frac{\pi \Phi}{\Phi_0}=\frac{\delta_2-\delta_1}{2}$

and $\varphi=\frac{\delta_2+\delta_1}{2}$

$I=I_{C_1}\sin{(\varphi-\frac{\pi \Phi}{\Phi_0})}+I_{C_2}\sin{(\varphi+\frac{\pi \Phi}{\Phi_0})}$

$I=I_{C_1}\exp{i(\varphi-\frac{\pi \Phi}{\Phi_0})}+I_{C_2}\exp{i(\varphi+\frac{\pi \Phi}{\Phi_0})}$

$I^2=I_{C_1}^2+I_{C_2}^2+2I_{C_1}I_{C_2}\cos{(\frac{2\pi \Phi}{\Phi_0})}$

$I^2=(I_{C_1}-I_{C_2})^2+4I_{C_1}I_{C_2}\cos^2{(\frac{\pi \Phi}{\Phi_0})}$

$|I|=\sqrt{(I_{C_1}-I_{C_2})^2+4I_{C_1}I_{C_2}\cos^2{(\frac{\pi \Phi}{\Phi_0})}}$


### -------------------------

```python
def I_standard(phi,Ic1,Ic2):
    return np.sqrt((Ic1-Ic2)**2+4*Ic1*Ic2*np.cos(np.pi*phi)**2)
```

```python
#Asymmetric

Ic1=0.8e-7
Ic2=0.9e-7

plt.plot(phi, I_standard(phi,Ic1,Ic2),label='asymmetric')
plt.ylim(0,1.8e-7)
plt.xlabel('$\Phi/\Phi_0$')
plt.ylabel('Current')

#Symmetric

Ic1=0.85e-7
Ic2=0.85e-7

plt.plot(phi, I_standard(phi,Ic1,Ic2),label='symmetric')
plt.legend()

#plt.plot(phi, I(phi,Ic1,Ic2))
```

What if for some reason $\delta_1$ is fixed at $\pi/2$ instead? (Low inductance limit?)

$I=I_{C_1}+I_{C_2}\sin{\delta_1}$

$\frac{2\pi\Phi}{\Phi_0}=\delta_2-\frac{\pi}{2}$

$I=I_{C_1}+I_{C_2}\sin{(\frac{2\pi\Phi}{\Phi_0}+\frac{\pi}{2})}$

$I=I_{C_1}+I_{C_2}\cos{(\frac{2\pi\Phi}{\Phi_0})}$


Assuming the current is entirely real (as opposed to the derivation above):


$|I|=|I_{C_1}+I_{C_2}(2\cos^2{(\frac{\pi\Phi}{\Phi_0})-1})|$



```python
def I(phi,Ic1,Ic2):
    return np.abs(Ic1+Ic2*(2*np.cos(np.pi*phi)**2-1))
```

```python
#Asymmetric

Ic1=0.8e-7
Ic2=0.9e-7

plt.plot(phi, I(phi,Ic1,Ic2),label='asymmetric')
plt.ylim(0,1.8e-7)
plt.xlabel('$\Phi/\Phi_0$')
plt.ylabel('Current')

#Symmetric

Ic1=0.85e-7
Ic2=0.85e-7

plt.plot(phi, I(phi,Ic1,Ic2),label='symmetric')
plt.legend()

#plt.plot(phi, I(phi,Ic1,Ic2))
```
